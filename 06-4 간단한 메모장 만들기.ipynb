원하는 메모를 파일에 저장하고 추가 및 조회가 가능한 간단한 메모장을 만들어 보자.
- 필요한 기능은? 메모 추가하기, 메모 조회하기
- 입력 받는 값은? 메모 내용, 프로그램 실행 옵션
- 출력하는 값은? memo.txt

memo.py는 우리가 작성할 파이썬 프로그램의 이름이다.

1. 우선 다음과 같이 입력으로 받은 옵션과 메모를 출력하는 코드를 작성해보자.

```python
# C:/doit/memo.py

import sys

option = sys.argv[1]
memo = sys.argv[2]

print(option)
print(memo)
```

sys.argv: 프로그램을 실행할 때 입력된 값을 읽어 들일 수 있는 파이썬 라이브러리
sys.argv[0]: 입력받은 값 중에서 파이썬 프로그램 이름인 memo.py 기능에는 필요없는 값이다.
sys.argv[1]: 프로그램 실행 옵션
sys.argv[2]: 메모 내용

2. memo.py를 작성했다면 다음 명령을 수행해 보자.


입력으로 전달한 옵션과 메모 내용이 그대로 출력되는 것을 확인할 수 있다.

3. 입력으로 받은 메모를 파일에 쓰도록 코드를 변경해보자.

```python
# C:/doit/memo.py

import sys

option = sys.argv[1]

if option == '-a':
    memo = sys.argv[2]
    f = open('memo.txt', 'a')
    f.write(memo)
    f.write('\n')
    f.close()
```



추가한 메모가 정상적으로 저장된 것을 볼 수 있다.

4. 이번에는 작성한 메모를 출력하는 부분을 만들 차례이다. 메모 출력은 -v 옵션을 사용한다.

```python
# C:/doit/memo.py

import sys

option = sys.argv[1]

if option == '-a':
    memo = sys.argv[2]
    f = open('memo.txt', 'a')
    f.write(memo)
    f.write('\n')
    f.close()
elif option == '-v':
    f = open('memo.txt')
    memo = f.read()
    f.close()
    print(memo)
```


명령 실행 시 입력한 메모가 그대로 출력되는 것을 확인
